```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("train_reports.csv")
df_test = pd.read_csv("test_reports.csv")
```

```python
for mode in list(df_train['mode'].unique()):
    df_temp = df_train[df_train['mode'] == mode]
    x_temp_wins = df_temp['wins_x'].to_numpy()
    o_temp_wins = df_temp['wins_o'].to_numpy()
    draws_temp = df_temp['draws'].to_numpy()
    rounds_temp = ['10','100','1000','10000','100000']
    plt.title(f"Evolução dos resultados ao longo do numero de partidas jogadas {mode} ")
    plt.bar(rounds_temp,x_temp_wins,color='b')
    plt.bar(rounds_temp,o_temp_wins,bottom=x_temp_wins,color='g')
    plt.bar(rounds_temp,draws_temp,bottom=x_temp_wins+o_temp_wins,color='r')
    plt.xlabel("Partidas usadas no treino")
    plt.ylabel("Resultados")
    plt.legend(["Ganhos (X)", "Ganhos (O)", "Velha"])
    plt.show()

```
